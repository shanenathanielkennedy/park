```python
## replaces long answers with short answers in former column
tutors['former'] = tutors['former'].replace(['Yes, I have tutored with GUTS Drop-In before.', 'Yes, I have tutored with GUTS before but not with Drop-In.'], 'yes')
tutors['former'] = tutors['former'].replace('No, I have not tutored with GUTS before.', 'no')

## maps modality preference to scale 1-5, with 0 being online only and 5 being in person only ((catch unexpected values later?)

tutors['preference'] = tutors['preference'].replace('Yes, I am only available to tutor in person', 5)
tutors['preference'] = tutors['preference'].replace('I prefer to tutor in person, but will tutor either in person or online', 4)
tutors['preference'] = tutors['preference'].replace('No preference', 3)
tutors['preference'] = tutors['preference'].replace('I prefer to tutor online, but will tutor either in person or online', 2)
tutors['preference'] = tutors['preference'].replace('Yes, I am only available to tutor online', 1)

# converts hrs_per_week to int and removes label
tutors['hrs_per_week'] = tutors['hrs_per_week'].str.slice(0, 1).astype(int) 
tutors
```

```python
#converts availability hours string into list of strings # maybe change to list comprehension
for column in tutors.columns:
    if column[:5] == "avail":
        tutors[column] = tutors[column].str.split(',')
```

```python
# convert availability
def format_hours(blocks):
    blocks1 = []
    for block in blocks:
        for hour in block:
            blocks1.append(hour)
    return list(set(blocks1))

def avail_str_to_int(df, column_name):
    for blocks in df[column_name]:
        if type(blocks) != float:
            for i in range(len(blocks)):
                blocks[i] = blocks[i].strip()
                if blocks[i] == "12:00-2:00pm":
                    blocks[i] = [12,1]
                elif blocks[i] =="2:00-4:00pm":
                    blocks[i] = [2,3]
                elif blocks[i] == "4:00-6:00pm":
                    blocks[i] = [4,5]
                elif blocks[i] == "5:00-7:00pm":
                    blocks[i] = [5,6]
                elif blocks[i] == "6:00-8:00pm":
                    blocks[i] = [5,6]
                elif blocks[i] == "7:00-9:00pm":
                    blocks[i] = [7,8]
                elif blocks[i] == "8:00-10:00pm":
                    blocks[i] = [8,9]
                else:
                    return 'hmm'
    return df
```
